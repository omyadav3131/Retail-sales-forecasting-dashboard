```python
import pandas as pd
#load dataset
df = pd.read_csv("../data/raw/train.csv")

print("Shape:", df.shape)

print("\nINFO:")
df.info()

print("\nFirst 5 Rows:")
print(df.head())

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
import numpy as np

#Replace 0 visibility with NaN
df['Item_Visibility'] = df['Item_Visibility'].replace(0,np.nan)

#check zero count
print("Zero count:",(df['Item_Visibility'] ==0).sum())

#check Missing values count
print("Missing count:",df['Item_Visibility'].isnull().sum())
```

```python
df['Item_Visibility'] = df.groupby('Item_Identifier')['Item_Visibility'].transform(lambda x: x.fillna(x.mean()))

print("Remaining missing values: ",df['Item_Visibility'].isnull().sum())
```

```python
df['Item_Weight'] = df.groupby('Item_Identifier')['Item_Weight'].transform( lambda x: x.fillna(x.mean()))
print("Remaining missing Item_Weight: ", df['Item_Weight'].isnull().sum())
```

```python
df['Item_Weight'] = df['Item_Weight'].fillna(df['Item_Weight'].mean())
print("Remaining missing Item_Weight: ", df['Item_Weight'].isnull().sum())
```

```python
df['Outlet_Size'].value_counts()
```

```python
df['Outlet_Size'] = df['Outlet_Size'].fillna('Medium')

print("Remaining Missing Outlet_Size: ", df['Outlet_Size'].isnull().sum())
```

```python
df['Store_Age'] = 2013 - df['Outlet_Establishment_Year']

df[['Outlet_Establishment_Year','Store_Age']].head()
```

```python
df['Item_Fat_Content'].unique()
```

```python
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace({
    'LF':'Low Fat',
    'low fat': 'Low Fat',
    'reg': 'Regular'
})
df['Item_Fat_Content'].unique()
```

```python
df.to_csv("../data/processed/cleaned_data.csv",index=False)
```
